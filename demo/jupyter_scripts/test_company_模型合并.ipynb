```python
def merge_category(data):
    if data.drop_duplicates().shape[0] == 3:
        return data["category_bilstm_resnet"]
    
    values, counts = np.unique(data, return_counts=True)
    return values[np.argmax(counts)]

df_merge["category_merge"] = df_merge[["category_xgboost", "category_lstm", "category_bilstm_resnet"]].apply(merge_category, axis=1)
df_merge
```
